P2.3 a, b, c:

```python
import math
import cmath

N = 1e28 #m^-3
w_0 = 6e15 #rad/sec
gamma = w_0 / 5 #damping
E_0 = 1e4 #V/m
e = 1.602e-19 #electron charge (C)
m = 9.11e-31 #electron mass (kg)

#part a code:

def chargeDisplacementData(angularFrequency):
    realTerm = w_0**2 - angularFrequency**2
    iTerm = gamma * angularFrequency
    r_e = (e * E_0) / (m * (realTerm**2 + iTerm**2)**(1 / 2))

    if (angularFrequency == w_0):
        phase = 90
    else:
        phase = math.atan(iTerm/realTerm) * (180 / math.pi)
        if (phase < 0):
            phase = (180 + phase)

    return (angularFrequency, r_e, phase)

#part b code:

epsilon_0 = 8.854e-12 #F/m
w_p = math.sqrt(N * e**2 / (epsilon_0 * m)) #rad/sec

def complexSusceptibilityData(angularFrequency):
    realTerm = w_0**2 - angularFrequency**2
    iTerm = gamma * angularFrequency
    susceptibilityMag = w_p ** 2 / (realTerm**2 + iTerm**2)**(1 / 2)

    if (angularFrequency == w_0):
        phase = 90
    else:
        phase = math.atan(iTerm/realTerm) * (180 / math.pi)
        if (phase < 0):
            phase = (180 + phase)

    return (susceptibilityMag, phase)

#part c code:

def complexRefractiveIndexData(angularFrequency):
    realTerm = w_0**2 - angularFrequency**2
    iTerm = gamma * angularFrequency

    realSusceptibility = (w_p**2 * realTerm) / (realTerm**2 + iTerm**2)
    iSusceptibility = (w_p**2 * iTerm) / (realTerm**2 + iTerm**2)

    Z = complex(1 + realSusceptibility, iSusceptibility)
    complexRefractiveIndex = cmath.sqrt(Z)

    realRefractiveIndex = complexRefractiveIndex.real
    iRefractiveIndex = complexRefractiveIndex.imag

    return (realRefractiveIndex, iRefractiveIndex)

#part i:

angularFrequency = w_0 - (2 * gamma)
iData = chargeDisplacementData(angularFrequency) + complexSusceptibilityData(angularFrequency) + complexRefractiveIndexData(angularFrequency)
print("Data for i:\n\nAngular Frequency: w = {point0:.2e} rad/sec".format(point0 = iData[0]))
print("Magnitude: r_e = {point1:.2e} meters\nPhase: ".format(point1 = iData[1]) + "phi = {point2:.2f} degrees".format(point2 = iData[2]))
print("Susceptibility Magnitude: |X(w)| = {point3:.2e}\nPhase: ".format(point3 = iData[3]) + "phi = {point4:.2f} degrees".format(point4 = iData[4]))
print("Real Refractive Index Component: n = {point5:.2f}\nImaginary Refractive Index Component: kappa = {point6:.3f}\n".format(point5 = iData[5], point6 = iData[6]))

#part ii:

angularFrequency = w_0
iData = chargeDisplacementData(angularFrequency) + complexSusceptibilityData(angularFrequency) + complexRefractiveIndexData(angularFrequency)
print("Data for ii:\n\nAngular Frequency: w = {point0:.2e} rad/sec".format(point0 = iData[0]))
print("Magnitude: r_e = {point1:.2e} meters\nPhase: ".format(point1 = iData[1]) + "phi = {point2:.2f} degrees".format(point2 = iData[2]))
print("Susceptibility Magnitude: |X(w)| = {point3:.2e}\nPhase: ".format(point3 = iData[3]) + "phi = {point4:.2f} degrees".format(point4 = iData[4]))
print("Real Refractive Index Component: n = {point5:.2f}\nImaginary Refractive Index Component: kappa = {point6:.3f}\n".format(point5 = iData[5], point6 = iData[6]))

#part iii:
angularFrequency = w_0 + (2 * gamma)
iData = chargeDisplacementData(angularFrequency) + complexSusceptibilityData(angularFrequency) + complexRefractiveIndexData(angularFrequency)
print("Data for iii:\n\nAngular Frequency: w = {point0:.2e} rad/sec".format(point0 = iData[0]))
print("Magnitude: r_e = {point1:.2e} meters\nPhase: ".format(point1 = iData[1]) + "phi = {point2:.2f} degrees".format(point2 = iData[2]))
print("Susceptibility Magnitude: |X(w)| = {point3:.2e}\nPhase: ".format(point3 = iData[3]) + "phi = {point4:.2f} degrees".format(point4 = iData[4]))
print("Real Refractive Index Component: n = {point5:.2f}\nImaginary Refractive Index Component: kappa = {point6:.3f}\n".format(point5 = iData[5], point6 = iData[6]))
```

Part a:

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath

%matplotlib inline

w_p = 1 #rad/sec
w_0 = 10 * w_p
gamma = 0.1 * w_p

def complexRefractiveIndexData(angularFrequency):
    realTerm = w_0**2 - angularFrequency**2
    iTerm = gamma * angularFrequency

    realSusceptibility = (w_p**2 * realTerm) / (realTerm**2 + iTerm**2)
    iSusceptibility = (w_p**2 * iTerm) / (realTerm**2 + iTerm**2)

    Z = complex(1 + realSusceptibility, iSusceptibility)
    complexRefractiveIndex = cmath.sqrt(Z)

    realRefractiveIndex = complexRefractiveIndex.real
    iRefractiveIndex = complexRefractiveIndex.imag

    return [realRefractiveIndex, iRefractiveIndex]

realIndexData = np.array([])
iIndexData = np.array([])
angular_frequency_range = np.array([])

for i in np.arange(0, 21, 0.1):
    angular_frequency = i * w_p
    complex_index = complexRefractiveIndexData(angular_frequency)

    realIndexData = np.append(realIndexData, complex_index[0])
    iIndexData = np.append(iIndexData, complex_index[1])
    angular_frequency_range = np.append(angular_frequency_range, (angular_frequency / w_p))

plt.plot(angular_frequency_range, realIndexData, label = 'Real Index vs Ang. Freq.')
plt.plot(angular_frequency_range, iIndexData, label = 'Imaginary Index vs Ang. Freq.')

plt.xlabel('w / w_p')
plt.ylabel('Refractive Index Value (Real and Imaginary)')
plt.legend()

print("N AND K AS A FUNCTION OF ANGULAR FREQUENCY FOR A DIAELECTRIC:")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath

%matplotlib inline

w_p = 1

w_0 = [10 * w_p, 15 * w_p, 25 * w_p]
gamma = [0.05 * w_p, 0.05 * w_p, 0.05 * (3 * w_p)]
f = [0.5, 0.25, 0.25]

def complexMRSusceptibility(resonant_frequencies, damping_constants, oscillator_strength):
    real_index_values = []
    imag_index_values = []
    angular_frequency_range = []

    for step in np.arange(0, 31, 0.1):
        oscillator_sum = 0
        angular_frequency = step * w_p

        for resonanceIndex in range(0, len(w_0)):
            oscillator_sum = oscillator_sum + (oscillator_strength[resonanceIndex] * w_p**2) / (resonant_frequencies[resonanceIndex]**2 - angular_frequency**2 - 1j * gamma[resonanceIndex] * angular_frequency)

        epsilon = 1 + oscillator_sum
        complex_index = cmath.sqrt(epsilon)
        real_index_values.append(complex_index.real)
        imag_index_values.append(complex_index.imag)
        angular_frequency_range.append(angular_frequency / w_p)

    return [angular_frequency_range, real_index_values, imag_index_values]

epsilon_values = complexMRSusceptibility(w_0, gamma, f)

plt.plot(epsilon_values[0], epsilon_values[1], label = "Real Index vs Ang. Freq.")
plt.plot(epsilon_values[0], epsilon_values[2], label = "Imaginary Index vs Ang. Freq.")

plt.xlabel('w / w_p')
plt.ylabel('Refractive Index Value (Real and Imaginary)')
plt.legend()

print("N AND K AS A FUNCTION OF ANGULAR FREQUENCY FOR A DIAELECTRIC GIVEN MULTIPLE RESONANT FREQUENCIES:")
plt.show()
```
